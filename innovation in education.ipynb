```python
# let us Compare Literacy rates across the US, China and Germany
US = dataframe[ dataframe.COUNTRY_Code == 'United States USA' ]
China = dataframe[ dataframe.COUNTRY_Code == 'China CHN' ]
Germany = dataframe[ dataframe.COUNTRY_Code == 'Germany DEU' ]
Germany
```

```python
plt.plot(US.Year, US.rates_of_Literacy)
plt.plot(China.Year, China.rates_of_Literacy)
plt.plot(Germany.Year, Germany.rates_of_Literacy)
plt.legend(['The US', 'China', 'Germany'])
plt.xlabel('Year')
plt.ylabel('rates_of_Literacy')
plt.show()
```
